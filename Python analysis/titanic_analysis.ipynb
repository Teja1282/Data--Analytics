```python
# finding the outliers
def outlier(col):
    hvo = []
    lvo = []

    for i in col:
        q3 = df[i].describe()['75%']
        q1 = df[i].describe()['25%']
        iqr = q3 - q1

        for value in df[i]:
            if value > (q3 + 1.5 * iqr):
                hvo.append((i, value))  
            elif value < (q1 - 1.5 * iqr):
                lvo.append((i, value))
    ply.figure(figsize=(20,10))
    sns.boxplot(data=df[col])
    ply.title("Boxplot of Selected Columns")
    ply.show()
    return hvo, lvo


col = ['PassengerId','Survived','Pclass','Age','SibSp','Parch','Fare']
outlier(col)
        
    
```
